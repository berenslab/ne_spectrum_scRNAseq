```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
# load data
data = np.load("data/human-409b2/human-409b2.data.npy")
labels = np.load("data/human-409b2/human-409b2.labels.npy")
with open("data/human-409b2/human-409b2.colors.json", "r") as f:
    colors = json.load(f)
print(data.shape)
```

```python
fig, ax = plt.subplots(figsize=(3,3))
ax.scatter(*data[:, :2].T, c=[colors[label] for label in labels], s=1, edgecolors='none', alpha=0.5)
ax.set_aspect('equal')
ax.axis('off')
```
